```python
@app.route('/individual/<path:iri>')
# this is for indivoduals of everything (MIneral, Rock, Physical_Property, ...)
# the iri is sent to here from the class_page.
def individual_page(iri):
    individual = IRIS[iri]
    
    # After clicking on an individual, it will write the name of the indivdual. 
    html = """<html><body><h2> <ol>'%s'</ol> </h2>""" %individual.name
    
    html += """<b><h3><p style="color:red"><br><u><b>Instance of</b></u>:</p></h3></b>"""
    for Class in individual.is_a:
        html += """<p><a href="%s"><ol>%s</ol></a></p>""" %(url_for("class_page", iri = Class.iri), Class.name)
    
    
    # Get a list of properties of a given individual and associated values
    html += """<b><h3><p style="color:red"><br><u>Diagnostic Properties</u></p></h3></b>"""
  
    for prop in individual.get_properties():
        # note: for this to work, properties must be asserted to the individuals, e.g., it works for biotite
        #print ('these are the properties of the individual: ', prop)
        for value in prop [individual]:
            html += """ <b>%s</b> : %s <br>""" %(prop, value)
   
    html += """<b><h3><p style="color:red"><br><u>All Properties</u></h3></p></b>"""
    
    html += """</body></html>"""
    return html
          
# Werkzeug is a Python Module installed by Flask. It launches the Website    
import werkzeug.serving
werkzeug.serving.run_simple("localhost", 5000, app) # Once executed, the website can be consulted on the localhost.
```
